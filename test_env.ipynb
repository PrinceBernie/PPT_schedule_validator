```python
import pandas as pd
```

```python
schedule_df = pd.read_excel("simulated_schedule_file.xlsx")
```

```python
schedule_df
```

```python
def numeric_clean(numeric_field):
    cleaned = numeric_field.astype(str).str.strip()
    transformed = pd.to_numeric(cleaned)
    return transformed

def clean_id_field(id_field):
    return id_field.astype(str).str.replace("[^0-9a-zA-Z]", "", regex = True).str.replace(r"\s+", " ", regex=True).str.strip()

def clean_name_field(name_field):
    cleaned = name_field.astype(str).str.replace("[^a-zA-z]", " ", regex = True).str.replace(r"\s+", " ", regex=True).str.strip().str.lower()
    return cleaned.apply(lambda x: " ".join(sorted(x.split())))

def rename_schedule_columns(schedules_df):
    schedule_df.columns = ("SSNIT Number", "NIA Number", "Contact", "Scheme Number", "Member Name", "Basic Salary", "Tier2")
    return schedule_df

```

```python
schedule_df = rename_schedule_columns(schedule_df)
schedule_df["SSNIT Number"] = clean_id_field(schedule_df["SSNIT Number"])
schedule_df["NIA Number"] = clean_id_field(schedule_df["NIA Number"])
schedule_df["Contact"] = numeric_clean(schedule_df["Contact"])
schedule_df["Scheme Number"] = clean_id_field(schedule_df["Scheme Number"])
schedule_df["clean_name"] = clean_name_field(schedule_df["Member Name"])
schedule_df["Basic Salary"] = numeric_clean(schedule_df["Basic Salary"])
schedule_df["Tier2"] = numeric_clean(schedule_df["Tier2"])
```
